```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
```

```python
df.columns
```

```python
vars = ['SMQ020', 'RIAGENDR', 'RIDAGEYR',
        'BMXWT', 'BMXHT', 'BMXBMI','BMXWAIST', 'HIQ210']
```

```python
df = df[vars].dropna()
df
```

```python
df.corr
```

## BMXWT = b0 + b1*RIDAGEYR

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR', data = df)
result = model.fit()
result.summary()
```

```python
df['RIDAGEYR'] = df.RIDAGEYR.replace({1:'Male', 2:'Female'})
df
```

### BMXWT = b0 + b1*RIDAGEYR + b2*RIAGENDR
### BMXWT = b0 + b1*RIDAGEYR + b2*(Male + Female)
### Female, categorica 1, categorica 2
### BMXWT = 77.5710 + -0.0228*RIDAGEYR + 10.5371*Male

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR + RIAGENDR', data = df)
result = model.fit()
result.summary()
```
