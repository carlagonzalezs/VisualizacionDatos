```python
df = pd.read_excel('datasets/Canada.xlsx',
                   sheet_name = 'Canada by Citizenship',
                   engine = 'openpyxl',
                   skiprows = range(20),
                   skipfooter = 2
                  )
df
```

```python
df.columns
```

```python
all(isinstance(column,str) for column in df.columns)
```

```python
df.columns = list(map(str, df. columns))
all(isinstance(column,str)for column in df.columns)
```

```python
df.drop(['Type','Coverage', 'REG', 'DEV','Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],
          inplace = True,
        axis = 1)
df.columns 
```

```python
df.rename(columns = {'OdName': 'Country',
                    'AreaName': 'Continent',
                    'RegName': 'Region',
                    },
         inplace = True)
df.columns
```

```python
df['Total'] = df.iloc[:,5:].sum(axis = 1)
df.head()
df.loc['Japon',5:8]
```

```python
df.loc['Japan','Total']
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
df.loc['Japan',years].sum
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('ggplot')
```

```python
df.loc['Japan', years].plot()
```

```python
print(plt.style.available)
```

```python
df.loc['Haiti', years].plot()
plt.xlabel('years', fontsize = 16)
plt.xticks(fontsize = 16, rotation = 45)

plt.ylabel('Numero de inmigrantes', fontsize = 16)
plt.yticks(fontsize = 16, rotation = 45)

plt.title('Numero de inmigrantes desde Haiti a Canada desde 1980-2013', fontsize = 13)
plt.text(20,6000, 'Terremoto')
plt.show()
```

## conclusion 

La inmigracion desde 1980 hasta 2010 fue un rango desde 1500-3500, pero en 2009 fue una escala
