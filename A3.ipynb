```python
my_pivot = pd.pivot_table(df,
                          index = ['desc_ent'],
                          values = ['peso'],
                          #aggfunc = ['std','sum','mean','median']
                          aggfunc = 'mean',
                          margins = True
                         )
my_pivot
```

```python
df = pd.read_csv('datasets/adult.csv')
df.head()
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['education','marital-status'],
                          values = 'age',
                          columns = 'sex',
                          aggfunc = 'mean',
                          margins = True
                         )
my_pivot.head(61)
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['education','marital-status'],
                          values = 'age',
                          columns = 'sex',
                          aggfunc = 'median',
                          margins = True
                         )
my_pivot.head(20)
```
