# 6 ¿Cuál es el porcentaje de la cuenta de cada mes y los cuartos del año?

```python
counts = df['Quarter'].value_counts()
type(counts)
counts = counts.to_frame() #para hacerlo una tabla
#print(type(counts))
#print(counts)
counts['%'] = counts.loc[:,'Quarter']/counts.loc[:,'Quarter'].sum(axis = 0)*100
counts.sort_index()
```

```python
counts['Quarter'].plot(kind = 'pie',figsize=(5, 6),
                            autopct='%10.1f%%', 
                            startangle=90,     
                            shadow=True,        
                            )
                      
plt.title('Total de casos que se realizaron en cada cuarto de año')
plt.axis('equal') 

plt.show()
```

## Conclusión: Se observa como hay más porcentaje de casos diarios en el Q3 del año, que es de Julio a Septiembre.


# 7 ¿Cuál es el impacto de Daily Cases en Month ID?
### Daily Cases = b0 + b1*Month ID 

```python
lm = linear_model.LinearRegression()
lm
```

```python
x = df['Daily Cases']
y = df['Month ID']
x = x.to_frame()
y = y.to_frame()
type(x)
```

```python
model_lm = lm.fit(y,x)
```

```python
model_ols = sm.OLS.from_formula('x~y', data = df) #height = b0 + b1*weight;  weight = b0 + b1*height;
result = model_ols.fit()
result.summary()
```

## Observaciones
### Variable dependiente: Month ID
### Variable independiente: Daily Cases 
### R squared: 0.017, por lo tanto el modelo no es tan eficiente, por que explica solo el 1.7% de Month ID 

### Df Model: 1
### intercept: >0.05, no es estadisticamente significativo

` Daily Cases = 4891.3857 + 220.1810*Month ID
