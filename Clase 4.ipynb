```python
sns.lmplot (x = 'Wingspan',
           y = 'CWDistance',
           data = df,
            fit_reg = False,
            hue = 'Gender'
           )
plt.title('Scatter plot of Wingman vs CWDistance', fontsize = 17)
plt.xticks(fontsize = 15)
plt.show()
```

## Conclusión:
no se encontro ninguna correlación entre a y b
