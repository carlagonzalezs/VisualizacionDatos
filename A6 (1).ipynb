## Mis variables de interés son years, month ID, day, Accum Cases y Daily Cases 


### 'years' es una variable númerica, donde se muestra el año en que se clasifica cada dato de los casos diarios de Covid 19

### 'month ID' es una variable númerica, donde se muestra el mes del dato del caso diario de covid 19 

### 'day' es una variable númerica, donde se muestra el día del dato del caso diario de covid 19 

### 'Accum Cases' es una variable númerica, donde es la suma de la acumulación de los casos diarios de covid, conforme pasan los días

### 'Daily Cases' es una variable númerica, donde explica la cantidad exacta por día de los datos de covid 19 

```python
df.drop(['Quarter ID', 'Quarter','Month','Time ID',
       'Time', 'Nation ID', 'Nation','AVG 7 Days Accum Cases', 'AVG 7 Days Daily Cases', 'Rate Daily Cases',
       'Rate Accum Cases', 'Days from 50 Cases', 'Style'],
       inplace = True,
        axis = 1
       )
df.columns
```

```python
import seaborn as sns
```

```python
sns.histplot(df["Daily Cases"], 
             kde = False).set_title("Histogram of Daily cases")

# kernel density estimation  KDE

plt.show()
```

```python
sns.boxplot(x = df["Month ID"], 
            whis=(0, 100)).set_title("Box plot of cintura")
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Month ID'],
                        values = ['Daily Cases'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'sum',
                        columns = 'Year',
                        margins = True
                         )
            
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Month ID'],
                        values = ['Daily Cases'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'median',
                        columns = 'Year',
                        margins = True
                         )
            
my_pivot.plot()
```

```python
sns.histplot(df["Accum Cases"], 
             kde = False).set_title("Histogram of Accum Cases")

# kernel density estimation  KDE

plt.show()
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Month ID'],
                        values = ['Daily Cases','Year'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'median',
                        margins = True
                         )
            
my_pivot.plot.barh()
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Year'],
                        values = ['Daily Cases'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'mean',
                        margins = True
                         )
            
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['Year'],
                        values = ['Daily Cases'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'mean',
                        margins = True
                         )
            
my_pivot.plot.barh()
```
