```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(30)
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['desc_ent'],
                        values = ['peso'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'median',
                        margins = True
                         )
            
my_pivot
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['desc_ent'],
                        values = ['peso'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'median',
                        columns = 'sexo',
                        margins = True
                         )
            
my_pivot
```

```python
df = pd.read_csv('datasets/adult.csv')
df.head(20)
```

```python
df.columns
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['education','marital-status'],
                        values = ['age'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'median',
                        columns = ['sex'],
                        margins = True
                         )
            
my_pivot.head(61)
```

```python
my_pivot = pd.pivot_table(df,
                        index = ['education','marital-status'],
                        values = ['hours-per-week'],
                        #aggfunc = ['std','sum','mean','median'],
                        aggfunc = 'mean',
                        margins = True
                         )
            
my_pivot.head(61)
```
