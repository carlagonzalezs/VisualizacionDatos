```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_excel('datasets/mic.xlsx',
                  engine = 'openpyxl',
                  )
df.head()
```

```python
counts = df['Age'].value_counts()
type(counts)
counts = counts.to_frame() #para hacerlo una tabla
#print(type(counts))
#print(counts)
counts['%'] = counts.loc[:,'Age']/counts.loc[:,'Age'].sum(axis = 0)*100
counts
```

```python
counts = counts.reindex(['<=20', '21 - 30', '31 - 40', '41 - 50'])
counts
```

```python
counts = counts.reset_index()
counts
```

```python
counts = counts.rename({'index': 'Age_group'},
                      axis = 1)
counts
```

```python
counts['Age'].plot(kind = 'pie')
```

```python
color_list = ['gold','yellowgreen', 'coral', 'skyblue']
explode_list = [0.03, 0.04, 0.06, 0.05]

counts['Age'].plot(kind = 'pie',
                  figsize = (15,6),
                   startangle = 90,
                   labels = None,
                   autopct = '%1.1f%%',
                   pctdistance = 1.1,
                   colors = color_list,
                   explode = explode_list,
                   shadow = True
                  )
plt.legend(labels = counts.Age_group,
           title = 'Legend',
           title_fontsize = 12,
          bbox_to_anchor = (1.0, 0.9)
          )
plt.title('Piechart of Age_group', fontsize = 15)
```

```python
x = list(counts['Age_group'])
y = list(counts['Age'])

plt.bar(x,y, width = 0.7, color = 'darkcyan')
plt.title('GrÃ¡fica de barras de Age_group', fontsize = 13)
plt.show()
```

```python
plt.subplot(1,2,1)

x = list(counts['Age_group'])
y = list(counts['Age'])

plt.bar(x,y, width = 0.7, color = 'darkcyan')
plt.title('GrÃ¡fica de barras de Age_group', fontsize = 13)


plt.subplot(1,2,2)

color_list = ['gold','yellowgreen', 'coral', 'skyblue']
explode_list = [0.03, 0.04, 0.06, 0.05]

counts['Age'].plot(kind = 'pie',
                  figsize = (15,6),
                   startangle = 90,
                   labels = None,
                   autopct = '%1.1f%%',
                   pctdistance = 1.1,
                   colors = color_list,
                   explode = explode_list,
                   shadow = True
                  )
plt.legend(labels = counts.Age_group,
           title = 'Legend',
           title_fontsize = 12,
          bbox_to_anchor = (1.0, 0.9)
          )
plt.title('Piechart of Age_group', fontsize = 15)
plt.savefig('datasets/Age_group.png')


plt.show()
```

```python
age_genero = pd.pivot_table(index = ['Age', 'Sex'],
                           data = df,
                           aggfunc = len,
                           values = 'Availability')
age_genero
```

```python
counts = df.groupby(['Age','Sex']).size()
counts = counts.to_frame() #para hacerlo una tabla
counts['%'] = counts.loc[:,0]/counts.loc[:,0].sum(axis = 0)*100
counts = counts.rename(columns = {0: 'Counts'})
counts
```

```python
category_plot = sns.catplot(x = 'Age',
                           hue = 'Sex',
                           data = df,
                           kind = 'count',
                           )

plt.xlabel('Age_group', fontsize = 17)
plt.xticks(fontsize = 17)
plt.title('Age_group por Sex', fontsize = 13)
```

```python
fig = px.sunburst(counts, 
                 path = ['Age', 'Sex'],
                  values = 'Counts',
                 )
fig.show()
```
